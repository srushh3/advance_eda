```python
df['proposed_date_of_completion'] = pd.to_datetime(df['proposed_date_of_completion'])
df['extended_date_of_completion'] = pd.to_datetime(df['extended_date_of_completion'])
df['revised_proposed_date_of_completion'] = pd.to_datetime(df['revised_proposed_date_of_completion'])
```

```python
df.head()
```

```python
sns.histplot(data=df, x='proposed_date_of_completion')
```

1. Majority of the flats are proposed to be complete between 2015 to 2030
2. around 3000 flats are proposed to be completed in 2025 and 2026

```python
sns.scatterplot(data=df, x='proposed_date_of_completion', y='revised_proposed_date_of_completion')
```

1. most of the buildings proposed to be completed between 2008 to 2026 are revised to be completed by 2018 to 2028
2. the data points have a strong relation
3. some outliers are present
4. the points slope upward from left to right so it has positive relationship between the two variables, thus they are directly related.


```python
# Create a scatter plot of extended date of completion vs proposed date of completion
sns.scatterplot(x='proposed_date_of_completion', y='extended_date_of_completion', data=df)
plt.show()
```

1. most of the buildings proposed date scheduled from 2012 to 2024 and further has been extended to 2018-2030
